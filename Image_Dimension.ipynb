```python
from google.colab.patches import cv2, cv2_imshow
img = cv2.imread('pinguim.png')
cv2_imshow(img) # Original image
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # Original to grayscale
cv2_imshow(img_gray)
ret, thresh = cv2.threshold(img_gray, 70, 255, 0) # Grayscale to binary
cv2_imshow(thresh)
```
