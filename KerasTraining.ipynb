```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

dataset = loadtxt('pima-indians-diabets.csv', delimiter=',')
x = dataset[:,0:8] # Pega os indicadores como idade, número de vezes que ficou grávida, pressão arterial, imc
y = dataset[:,8] # Resposta binária para diabets, 1 é sim e 0 é não

model = Sequential()
model.add(Dense(12, input_shape=(8,), activation='relu' ))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x, y, epochs=150, batch_size=10)

_, accuracy = model.evaluate(x, y)
print("Accuracy: %.2f" % (accuracy*100))

predictions = (model.predict(x)>0.5).astype(int)

for i in range(5):
  print("%s => %d (expected %d)" % (x[i].tolist(), predictions[i], y[i]))
```
